```python
#파일 종류: 바이너리(숫자-2진수가 저장된 파일)
# 바이너리 파일: 이미지, 동영상, 음성, 엑셀
# 텍스트 파일. 문자로 처리 (인코딩, 디코딩)
# 메모장에서 텍스트 파일을 열면 화면에 문자가 보인다(디코딩)

# 파일 생성 : 모드 (텍스트, 바이너리) 둘중에 결정
# 읽기(r), reset해서 쓰기(W), 추가쓰기(a)
# 파일 생성은: text 모드에서 덮어쓰기를 하려고함
# 덮어쓰기는 두가지 존대    (해당파일 있거나 없거나)
# 해당파일이 없을때는 새로 생성됨
```

```python
#파일 생성: 텍스트 모드, w, 해당 파일 없는 경우(생성)
#텍스트 파일을 생성 -> 파일명: 새파일, 확장자: txt, 위치: chap04
#파일 처리는 운영체제를(OS) 를 따른다

#순서
#1. open
#2. write
#3. close

# 상대경로 방식
# 여기서 open은 function임
f = open(
    '새파일.txt', #file 매개변수 값을 지정, 경로(경로를 명시 안하면 현재 디렉토리를 명시)
    # 와 파일명, 확장자(파일의 성격이나 용도를 나타냄)  
    'w'         # 텍스트, 쓰기 모드 지정
)

#open에 의해서 생성된 object
f.close() #메소드 

# 절대경로 방식
# f = open(
#     'C:\big21\python_dev\chap04\새파일.txt',
#     'w'     
# )
```

```python
#파일 읽고 쓰기 
f = open(file='./새파일.txt', mode = 'w', encoding='utf-8')

#10줄을 파일에 쓰기
for i in range(1,11):
    data = f'{i}번째 줄입니다.\n' #파일에 쓰여질 데이터 생성
    f.write(data)  #데이터를 파일에 써라
f.close()

```

```python
#파일 읽기
f = open('./새파일.txt', 'r', encoding='utf-8')
line = f.readline()  #한줄만 읽기
print(line)
f.close()
```

```python
#여러줄 읽어서 출력
f= open('./새파일.txt', 'r', encoding = 'utf-8')
for i in range(10):
    line = f.readline()
    print(line)

f.close()
```

```python
#readlines(): 한줄씩 읽고 전체를 리스트 반환
#리스트가 나온다는건 for문을 돌릴 수 있다는것  -> break 안써도되
#['1번째줄', '2번째줄', '3번째줄', '4번째줄',,,'9번째줄', '10번째줄']

#제일 많이 쓰는건 readlines
f = open('./새파일.txt', 'r', encoding = 'utf-8')
lines = f.readlines() #전체 읽기 처리 된거임 -> 리스트로 반환
print(lines)
for line in lines:
    print(line)
f.close()
```

```python
#read() : 파일 내용 전체를 통으로 문자열 반환 string반환
f= open(
    './새파일.txt', #읽을 파일의 경로(위치)와 파일명 지정
    'r',           # 파일을 읽기 위해 오픈
    encoding = "utf-8" #인코딩에 사용할 문자셋을 유니코드로 지정    
)
data = f.read()
f.close()
data.split('\n')
```

```python
#파일에 새로운 내용 추가
#모드: 'a'   append 추가 모드
f = open(
    './새파일.txt',
    'a', #추가모드
    encoding='utf-8'
)
for i in range(11,20):
    data = f'{i}번째 줄입니다.\n'
    print(data)
    f.write(data)
f.close()
```

```python
#자동으로 close()를 해주면
# 파일 새로 생성: foo.txt
f = open(
    './foo.txt',
    'w',
    encoding = "utf-8"
)
f.write('Life is too short')
f.close()
```

```python
# 자동 close
with open('./foo.txt', 'w', encoding = 'utf-8') as f:
    f.write("Life is too short")
```

```python
# 함수: 특정 작업을 수행하는 코드 묶음에 이름을 붙여, 필요할때마다 불러 쓸수 있게 만든것
# csv, JSON 파일에 대한 처리 입출력
# csv 파일 읽기, 쓰기 하는 기능(함수) 을 가지고 있는 모듈: csv
# csy.py (외장 모듈)
# 외장 모듈: 설치는 되어 있는데 외장이다
import csv

f = open("./data/data.csv", "r", encoding="utf-8")

rdr = csv.reader(f) #csv 형태로 읽을 수 있게 만든거

for line in rdr:
    print(line)

f.close()
#모듈안에는 GV, function, class가 존재
```

```python
# csv 파일에 쓰기
import csv
# data 디렉토리에 output.csv 파일 생성하고 쓰기
f= open('./data/output.csv', 'w', encoding= 'utf-8')
#csv write할 수 있는 object를 생성해야함
wr = csv.writer(f)
wr.writerow([1, '홍길동', False])
wr.writerow([2, '고길동', True])
f.close()
```

```python
# JSON파일 입출력
# 자바스크립트에서 오브젝트를 표기하는 문법: {key:value,K:v,,,,}
# JSON -> DICT -> JSON 로 바꿔주는 파이썬의 모듈: json module
# jsoon module 중 많이 쓰는 function: load(), dump()
```

```python
#json 쓰기: 파이썬 객체(dict) -> json 형태의 파일 저장
import json

데이터 = {
    '이름': '홍길동',
    '나이': 30,
    '취미':['코딩', '운동']
}

#json형식으로 파일에 저장  dump()
with open('./data/프로필.json', 'w', encoding = 'utf-8') as f:
    json.dump(
        데이터, #파일에 저장할 딕셔너리를 지정
        f,     # 어느 파일 
        ensure_ascii=False,  #한글이 깨지지 않고 그대로 저장시키기 위해서
        indent=2    #사람이 읽기 좋은 형태로 들여쓰기한다
    )
```

```python
#json 형식의 파일 -> 파이썬 객체 (딕셔너리) 읽기
with open('./data/프로필.json', 'r', encoding='utf-8') as f:
    dict_data = json.load(f)

print(dict_data.get('이름'))
print(dict_data.get('나이'))
print(dict_data.get('취미'))
```
