```python
#함수 선언과 실행
# 문제: 두정수를 입력받아 합을 구하는 함수 (입력 -> 코드 -> 반환)
# 입력: 두 정수
# 코드 : 코드는 함수의 처리 기능
# 반환: sum
# def 함수이름(입력):
#   코드
#   return
def sum_(a,b):
    result = a + b
    return result

def sum__(a, b):
    return a + b

print(sum__(3, 7))
```

```python
# 가변 인자 (arguments)
# 3개 정수의 합을 구하는 함수 선언, 실행
#똑같은 이름의 함수가 존재할 수 가 없다 
#다른 함수명으로 하면됨

#2개 들어올때
def 합계2(x,y):
    total = x + y 
    return total

#3개 들어올때
def 합계3(x, y, z):
    return x+y+z

합계2(1,2)
```

```python
# 가변 인자 적용
def 합계(*numbers): # *numbers : 가변 매게변수, 인수를 튜플(인덱스, 반복)로 처리
    total = 0
    for number in numbers:
        total += number
    return total

print(합계)
```

```python

print(합계(1,2))
print(합계(1,2,3))
print(합계(1,2,3,4))
```

```python
# 가변인자: **  -> 개수가 정해지지 않은 키워드 인자, 함수 내부에서 딕셔너리 형태로 처리된다
def 프로필(**정보): # ** 두개이면 딕셔너리로 처리하겠다의 의미
    for key, value in 정보.items():
        print(f"{key}, {value}")
```

```python
#변수의 유효 범위 (scope)
#scope가 실행되면 변수가 사라짐   
# 함수 안에서 만들 변수는 함수 안에서만 살아있다, 지역 변수라고 한다
# 4가지 스코프
# 1. local scope : 함수 내부 변수
# 2. Enclosing(중첩함수)
# 3. global
# 4. built-in

# LEGB rule임 -> 파이썬은 변수를 만나면 다음 순서로 찾는다
# local -> enclosing -> global -> built-in

```

```python
# 함수의 문서화 : 독스트링(docstring), 함수에 대한 설명 적어 놓은 것
# 환률변환 함수
# 구글 스타일의 docstring

def 환율변환(원화:float, 환율:float) -> float:  #float 타입힌트
    """원화를 외화로 환산한다.

    Args:
        원화: 변화할 금액(원)
        환율: 1외화당 원화 가격

    Returns: 
        변환된 외화 금액    
    """

    return 원화/환율
```

```python
print(환율변환.__doc__)
help(환율변환)
```

```python
#타입힌트 : 매게변수와 반환값이 어떤 자료형인지 표시하는 기능
def 평균(점수목록: list[int]):     # 점수목록 = [10, 9, 6, 4]
    return sum(점수목록) / len(점수목록)

평균 ([10, 9, 6, 4])
```
