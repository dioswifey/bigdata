```python
# 조건문: if, switch, match ~ case
# 기본적으로는 if문을 사용한다, 복잡해지면 match를 고려한다
# if를 아는게 가장 중요함

# 1. if 구조  : 특정 조건에 만족할 경우에만 어떤 코드 실행하고자 할때 
#   if 조건식:
#       조건식이 TRUE 일때 실행할 코드 블록
# 2. if ~ else 구조
#    if 조건식 :
#       조건식이 True일때 실행할 코드 블록
#    else:
#       조건식이 false일때 실행할 코드 블록
# 3. if ~ else  if ~ else 구조
```

```python
#돈이 있으면 택시를 타고, 돈이 없으면 걸어간다
#를 코드로 바꿔보자

#첫번째는 돈을 변수로 만들어야함
#왜냐하면 돈을 저장을 해야지 돈이 있나 없는지 알 수 있기 떄문
#돈은 숫자니까  int data type으로 저장해야한다
#정수를 저장할 변수를 선언해야한다

#변수명 선언은 영어로 해라, 카멜 표기법을 사용해라
#바이브 코딩할때  표기법을 얘기해줘야함
# 돈을 저장할 변수 필요
money = 1   #money 가 1이라는건 돈이 있다라는 뜻임ㅁ
 
#조건식
#colon의 기능 : block의 시작을 의미 block의 의미 : 여러 명령을 묶는 역할
#block 의 usage: 여러 명령을 할거 아니면 block을 쓸 필요가 없다 
if money : #money 가 true 일때의 block
    print("택시를 타라")
else: # money 가 false 일때의 block
    print("걸어가라")

#조건에 따라 조건이 true 또는 조건이 false


```

```python
# 만약 5000원 이상의 돈을 가지고 있으면 택시를 타고 그렇지 않으면 걸어가라

money = 5230

if money >= 5000:
    print("택시를 타라")
else:
    print("걸어가라")
```

```python
#돈이 5000원 이상있고 카드가 있으면 택시를 타고 아니면 걸어가라

#비교연산자만 갔고 해결할 수는 없음
#돈이 5000원이상 파트 : 비교1
#카드가 있으면 : 비교2
# 둘중에 하나만 있어도  ->  택시를 타라
# 둘중에 하나  라는 부분이 논리연산자 

#논리연산자 종류:  or, and , not
# 이 경우에서는 " 둘중 하나가 있거나"   -> 비교
card = 1
money = 3000

if money >= 5000 or card:
    print("택시를 타라")
else: 
    print("걸어가라")
```

```python
#in 연산자 => 특정값 in iterable(반복가능한 객체) 사용:
# x in list
# x in tuple
# x in string

# 여기서 list, tuple, stirng 3개 전부 다 iterable
# 여기서 iterable 은 여러개 저장하고 있는 놈이라고 생각하면된다 + 반복 가능 + index를 가지고 있다 

#in 은 조건식에 사용할 수 있다
#왜냐하면 true랑 false가 나올 수 있으니까
#meaning!! if문에 사용가능

#list
print(1 in [1,2,3])

#tuple과 string 즉 3개의 iterable 모두 다 가능

```

```python
# 주머니에 돈이 있으면 택시를 타고, 없으면 걸어가라

pocket = ["paper", "cellphone", "money"]


if "money" in pocket:
    print("택시를 타라")
else:
    print("걸어가라")
```

```python

#elif = 앞 조건이 실패했을 때 다른 조건을 추가 검사

pocket = ["card", "paper", "cell"]
if "money" in pocket:
    print("택시타고 가라")
elif "card" in pocket:
    print("택시타고 가라")
else:
    print("걸어가라")
```
