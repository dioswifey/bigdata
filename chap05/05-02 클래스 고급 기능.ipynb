```python
# 요건
# 박씨 관리하는 집: 성이 "박"
# fullname: 성 + 이름
# 여행 메소드: 장소를 입력하면 => "성 + 이름, 장소 여행을 가다"


class HousePark:
    lastname = "Park"   #클래스 변수   즉 공유 가능하다/ 클래스 변수가 있을땐 self가 있어야함

    def __init__(self, name):           #왠만하면 __init__ 쓰는게 좀 깔끔함
        self.fullname = self.lastname + name

    # def setName(self, firstname):
    #     self.fullName = self.lastName + firstname

    def travel(self, where):
        print(f"{self.fullname}: {where} 여행을 가다.")
```

```python
aa = HousePark("Gina")
aa.travel("Busan")
```

```python
#클래스의 상속
#상속의 목적은 추가하는것
#핸드폰에다가 스마트폰의 애플페이기능을 추가한다는것이 상속의 목적

# HousePark을 상속받는 HouseKim
# __init__ 이 없는 상태
class HouseKim(HousePark):
    lastname= "김"
    
```

```python
juliet = HouseKim("줄리엣")

#이부분은 단순 가져다쓴거임, 상속받은거 (OVERRIDING)
juliet.travel("서울")
```
