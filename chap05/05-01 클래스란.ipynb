```python
#클래스: 비슷한 성격의 데이터(변수)와 기능(메소드)를 하나로 묶어,
#오브젝트를 찍어내는 설계도(틀) 이다.
#설계도대로 생성된 놈들 -> 객체: 오브젝트, 인스턴스
#반복 사용 의미가 있다 : 함수 비슷하다

#왜 사용하는가?
#에를 들어 학생 한명의 정보를 저장: 
#이름, 점수

```

```python
#한명의 학생 정보를 저장하는 클래스 선언
class 학생: #학생이라는 클래스 선언해라 명령
#   이름, 점수 저장하는 공간 선언
    def __init__(self, 이름, 점수): #__init__() : 오브젝트가 생성될때 자동으로 호출 되는 메소드
        self.이름= 이름  #self.이름: 이름 변수 선언, 이름이라는 값을 대입하라
        self.점수= 점수  #self.점수: 오브젝트 안에 점수 변수 선언, 점수라는 값을 대입
    def 출력(self): #오브젝트 안에 값을 출력 메소드
        print(f"{self.이름} : {self.점수}")
```

```python
홍길동 = 학생("홍길동", 90)  #학생이라는 클래스 () 가 붙으면 오브젝트 생성하라는 명령
고길동 = 학생("고길동", 80)
마이꼴 = 학생("마이꼴", 70)
```

```python
#클래스를 쓰는 이유
#1. 데이터(변수)와 기능(메소드)를 묶기
#2. 재사용: 설계도 하나로 객체를 무한 생산
#3. 확장성: 상속으로 기존 코드를 물려받아 확장 가능
#4. 현실 모델링 : 현실의 사물을 자연스럽게 표현 => 자동차, 스마트폰
```

```python
# 계산기 클래스
class Calculator:
    def __init__(self):
        self.result = 0
    def adder(self,num):   #adder에 왜 self가 필요하냐? self.result = 0 에 result에 갈거라서
        self.result += num    #result에 num을 누적
        return self.result
    #빼기 추가
    def sub(self,num) :
        self.result -= num
        return self.result
    
```

```python
cal1 = Calculator()
cal2 = Calculator()
```

```python
print(cal1.adder(3))
print(cal1.sub(2))
```

```python
#지금부터 계산기 클래스
#요건: 더하기 기능을하고, 결과를 반환한다
#클래스이름을 요건에 맞게 줘라

#1] 클래스 이름: Caluculator, 해당 클래스의 용도를 설명할 수 있는 이름
#2] 변수: 결과 저장, 반환
#3] 기능: method   더하는 기능
# 계산기가 화면에 딱 나왔을떄 0값을 갖고 찍혀야뎀 - __init__ - self가 들어가야됨

class Calculator:
    def __init__(self):
        self.result = 0
        
    def adder(self, num):
        self.result += num
        return self.result
    
```

```python
calA = Calculator()

calA.adder(3)
```

```python
# 학생 정보 저장하고 출력 클래스
# 학생 정보: 학번, 학과, 이름, 성별
# 출력: 학번, 학과, 이름

class Student:
    def __init__(self, grad, major, name, gender):
        self.grad = grad
        self.major = major
        self.name = name
        self.gender = gender
    def Info(self):
    
        #self.gender값이 "여자" 인지 "남자"인지 따라서 print로 출력
        if self.gender == "여자":

            print(f"{self.grad} | {self.major} | {self.name}(F)")
        else:

            print(f"{self.grad} | {self.major} | {self.name}(m)")


stu1 = Student(2028, "Business", "Daniel", "남자")
stu1.Info()
```

```python
#product class만들기
# name, price, category 를 저장(variable), "가전제품" == 배송비가 추가  라는 메시지를 붙여줌
# 가전, 의류, 식품

class Product:
    def __init__(self, name, price, category):
        self.name = name
        self.price = price
        self.category = category

    def shippingFee(self):
        if self.category == "가전제품":
            print(f"상품명: {self.name} | 상품가격: {self.price} | 카테고리 : {self.category} | 배송메시지: 배송비 추가됨")
        if self.category == "의류" or self.category == "식품":
            print(f"상품명: {self.name} | 상품가격: {self.price} | 카테고리 : {self.category} | 배송메시지: 배송비 무료") 

macbook = Product("맥북프로", 2000, "가전제품")
macbook.shippingFee()
coat = Product("코트", 150, "의류")
coat.shippingFee()
        
```

```python
#self 받는 메소드에서만 오브젝트 변수 선언 가능
class Service:
    secret = "영구는 외계인이다"

    def setName (self, name):
        self.name = name
    def sum(self, a, b):
        result = a + b
        print(f"{self.name}님, {a}랑 {b}가 더하면 {result} 입니다")
```

```python
Rosa = Service()
Rosa.setName("Tulisa")
Rosa.setName("Franklyn")
Rosa.sum(2,3)
```
