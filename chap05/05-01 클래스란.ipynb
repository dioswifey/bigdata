```python
#클래스: 비슷한 성격의 데이터(변수)와 기능(메소드)를 하나로 묶어,
#오브젝트를 찍어내는 설계도(틀) 이다.
#설계도대로 생성된 놈들 -> 객체: 오브젝트, 인스턴스
#반복 사용 의미가 있다 : 함수 비슷하다

#왜 사용하는가?
#에를 들어 학생 한명의 정보를 저장: 
#이름, 점수

```

```python
#한명의 학생 정보를 저장하는 클래스 선언
class 학생: #학생이라는 클래스 선언해라 명령
#   이름, 점수 저장하는 공간 선언
    def __init__(self, 이름, 점수): #__init__() : 오브젝트가 생성될때 자동으로 호출 되는 메소드
        self.이름= 이름  #self.이름: 이름 변수 선언, 이름이라는 값을 대입하라
        self.점수= 점수  #self.점수: 오브젝트 안에 점수 변수 선언, 점수라는 값을 대입
    def 출력(self): #오브젝트 안에 값을 출력 메소드
        print(f"{self.이름} : {self.점수}")
```

```python
홍길동 = 학생("홍길동", 90)  #학생이라는 클래스 () 가 붙으면 오브젝트 생성하라는 명령
고길동 = 학생("고길동", 80)
마이꼴 = 학생("마이꼴", 70)
```

```python
고길동.출력()

#self값을 바꾸게 하는게 변수명.출력()  으로 self값이 바뀐다
```

```python
#클래스를 쓰는 이유
#1. 데이터(변수)와 기능(메소드)를 묶기
#2. 재사용: 설계도 하나로 객체를 무한 생산
#3. 확장성: 상속으로 기존 코드를 물려받아 확장 가능
#4. 현실 모델링 : 현실의 사물을 자연스럽게 표현 => 자동차, 스마트폰
```

```python
# 계산기 클래스
class Calculator:
    def __init__(self):
        self.result = 0
    def adder(self,num):   #adder에 왜 self가 필요하냐? self.result = 0 에 result에 갈거라서
        self.result += num    #result에 num을 누적
        return self.result
    #빼기 추가
    def sub(self,num) :
        self.result -= num
        return self.result
    
```

```python
cal1 = Calculator()
cal2 = Calculator()
```

```python
print(cal1.adder(3))
print(cal1.sub(2))
```
