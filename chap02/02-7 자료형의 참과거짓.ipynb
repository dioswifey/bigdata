```python
# 조건식 자리에 데이터 타입이 들어가면 참/거짓의 역할을 한다
# 데이터 타입(변수)에 값이 있으면 참, 없으면(비었으면) 거짓 역할을한다.

# 조건식 자리라는 것이중요
#예를 들어  데이터가 있으면 처리, 없으면 넘어가기

#숫자
#기본 if 
#if 조건식 자리:
#      조건식이 True 일때 실행될 코드가 들어갈 자리
#else: 
#      조건식이 false일때 실행될 코드가 들어갈 자리 
num = 0
if num:
    print('0이 아니다')
else:
    print('0이다')
```

```python

#조건식 true, false  boolenan 처리하는것
#조건식 기준으로 true랑 false갈리는거임

list_ = [] #빈 리스트

if list_:
    print('There is something inside the list')
else:
    print('The list is empty')

list_

```
