```python
#왜 리스트가 필요하냐를 설명해줌
#변수를 너무 많이 선언하기 싫음
#1000개의 변수를 한번에 저장하려면? 예를들어
#리스트를 사용하면 됨

name1 = "김소현"
name2 =  "고길동"
name3 = "마이꼴"


#리스트를 사용함으로써 한번에 3개의 이름 모두 다 저장가능

names = ["김소현","고길동","마이꼴"]
print(names)

```

```python
# 리스트 생성, 사용
# 1부터 10사이의 홀수 저장: 1,3,5,7,9
# 리스트 생성: [] 대괄호
odds = [1,3,5,7,9]
print(odds[2])
```

```python
#어떤 걸 저장할 수 있는지?
a = [] # 빈 리스트, 빈값을 저장, a = list() 동일
b = [1,2,3] #정수 저장
c = ["Life", "is", "too", "short"] #여러개의 문자열을 한번에 저장 가능
d = [1,2, "life", "is" ] #정수와 문자열 
e = [1,2, ["Life", "is"]] #리스트 안에 리스트 저장
#리스트는 6개 모두의 데이터 타입을 저장 할 수 있음
#중요한 리스트의 기능: 아무거나 다 저장할 수 있다. 
```

```python
# 리스트의 인덱싱과 슬라이싱
# 인덱싱: 단일값 추출 []
# 슬라이싱: 여러개값 추출,[처음값: 끝값 -1]
a = [1,2,3]
print(a)
print(a[2])
print(a[-1])
```

```python
a  = [1,2,3, ['a', 'b', 'c']]
print(a[3][2])

```

```python
#슬라이싱
a = [1,2,3,4,5]
print(a[0:2])
print(a[:2])
#3,4,5
print(a[2:])
```

```python
#어떤 반에 3명의 학생이 있다. 이름과 국어점수를 저장해서 출력하는 프로그램 
students = [
    ["홍길동", "고길동", "마이꼴"], #이름 저장하는 리스트
    [80, 90, 60]

]

print(students[1][1])
print(students[1][2])
print(students[0][1])
```

```python
# 리스트 연산자
a = [1,2,3]
b = [4,5,6]
#두개 리스트 합치기 
print(a + b)
#리스트 반복
print(a * 3)
```

```python
# 리스트의 수정, 변경, 삭제
# 1. 하나의 값 (단일값:인덱싱) 수정
a = [1,2,3]
a[2] = 4
print(a)
```

```python
#리스트에서 연속된 범위의 값(슬라이싱) 수정
#a = [1,2,4]
# print(a[1:2])
# a[1]
a[1:2] = ["a", "b", "c"]   #슬라이싱
# a[1] = ["a", "b", "c"]    #인덱싱으로 넣으면 리스트가 통째로 들어간다 
print(a)
```

```python
#리스트에서 연속된 범위의 값(슬라이싱) 수정
#a = [1,2,4]
# print(a[1:2])
# a[1]
a[1:2] = ["a", "b", "c"]   #슬라이싱
# a[1] = ["a", "b", "c"]    #인덱싱으로 넣으면 리스트가 통째로 들어간다 
print(a)
```

```python
a = [1, 'a', 'b', 'c', 4]
# 리스트 요소 삭제   'a'랑 'b' 삭제하고픔
a[1:3] = []
print(a)
```

```python
# 리스트 요소 삭제 두번째 방법 : del 함수 사용
# 리스트의 요소들은 각각 object임
# del 함수는 object자체를 삭제할 수 잇음
del a[1]
print(a)
```

```python
#리스트 관련 메소드들
# 1. 리스트의 마지막 값 추가 : append(추가할값)
a = [1,2,3]
#맨 뒤에 4 추가
a. append(4)
print(a)
```

```python
#현재 리스트 안에 리스트 추가 : [1,2,3,4,[5,6]]
a.append([5,6])
print(a)
```

```python
#리스트 정렬 : sort
#sort라는 메소드는 오름차순 밖에 안된
a = [1,4,3,2]
a.sort()
print(a)
```

```python
#알파벳 순서로 정렬
a  = ["a", "c", "b"]
a.sort()
print(a)
```

```python
#리스트 뒤집기: reverse(), 정렬은 아니야
a = ["a", "c", "b"]
a.reverse()
print(a)
```

```python
# 위치 반환 : index(값) => 찾을 값이 있을 경우와 없을경우
# 찾을 값이 있을 경우: 해당 위치의 인덱스 값을 반환
# 없을 경우: 예외(오류) 발생

a = [1,2,3]
a.index(3)
```

```python
#리스트에 요소 삽입: insert(삽입할 위치, 삽입할 값)   무조건 두개의 값을 넣어줘야함

a = [1,2,3]
# 첫번재 자리에가 4를 삽입하고싶음 
a.insert(1,4)
print(a)
```

```python
#리스트 요소 제거 :remove (제거할 값)
a = [1,2,3,1,2,3]
#3일란 값을제거
a.remove(3)
print(a)
a.remove(3)
print(a)
```

```python
#리스트에서 특정값을 끄집어 내기: pop(X), pop(특정값)
# 여기서 X는 값이 없음을 의미한다
# pop(X) ; 마지막 값을 끄집어낸다
#pop()안에 아무 값도 안준경우  리스트의 마지막 값을 끄집어낸다 
a = [1,2,3]
a.pop()
print(a)
#특정값을 줄 수 도 잇음
a.pop(1)
```

```python
a = [1,2,3]
b = [4,5,6]
# print(a.extend(b))
b.extend(a)
print(b)
```
