```python
#정수 데이터 타입 int
#정수: 음의 정수, 양의 정수, 0 
a = 123 #123이라는 정수를 a에 대입한 상태, a로 123 이라는 정수를 사용
a = -123 #음의 정수 -123의 주소값을 a에 대입한 상태. 
a = 0 
```

```python
#실수 데이터 타입 float
a = 1.2 
a = -3.45
a = 0.0
a = 42400000000.0
a = 4.24E3 #4.24 => 424 (오른쪽 방향)
a = 4.24e-3 #4.24 => 0.0424 (왼쪽 방향)
```

```python
# 16진수 : 0~15의 값으로 숫자를 표기하는방법
# 한자리값으로 표기해야한다 : 0 부터 9까지는 한자리로 표기할 수 있음. 10은 두자리가됨. 10을 한자리로 바꿔주는 게 10 -> A, 11 -> B, 12 -> C, 13 -> D, 14 -> E, 15 -> F  
# 표기법: 0x로 시작해야한다

a = 0xAB #0x로 시작한다는 뜻은 A하고 B는 16진수임
# 2진수로 표기했을때 굉장히 길어지지만 16진수로 처리하면 줄어든다. 유니코드에서 16진수를 사용한다
a = 171 #0xAB(16진수) = 171(10진수)
```

```python
#연산자
# 사칙연산 (+, -, *, /)
a = 3 #a에 3을 대입한 상태
b = 4 
print(a+b)
print(a/b)
```

```python
#거듭제곱 연산자 : **
a = 3
b = 3
print(a**b) #3의 3승
```

```python
#나머지 구하는 연산자: %
a = 7
b = 3
print(7%3) 
```

```python
#나눗셈 후 두정수 중 더 작은값을 반환하는 연산자: // 
#나누기
a=7
b=4
print(a/b)
print(a//b) 

a = -7
print(a/b)
print(a//b) #조심해야함. -는 버리는게 아님. 두정수 간의 더 작은값을 반환한다
#a//b = -1.75 => -1과 -2사이임. 둘중 더 작은값을 반환한다. 

```
