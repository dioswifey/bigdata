```python
#튜플 vs 리스트  : 인덱스 존재  => 인덱싱, 슬라이싱 가능
#튜플은 불변이다, 리스트는 가변이다  (중요한 차이점임)
#튜플 생성 : 
t1 = () #빈 소괄호 => 빈 튜플 오브젝트 생성된다
t2 = (1,) #1이 들어있는 튜플 오브젝트 생성된다
t3 = (1,2,3) 
# 1,2,3은 튜플이다, 불변이다 ㄴ
print(t2)
```

```python
# 불변 확인: 삭제
t1 = 1,2,3,4
#소괄호 없이 ,라 구분되어 있는것도 튜플이다
del t1[5]  #1 지우라는 얘기
#삭제가 안됨, 불변이기 때문에

#튜플은 저장만 하는 리스트, 수정을 못함 
#그럼 왜 튜플을 만들었을까?
```

```python
#튜플은 인덱스를 가지고 있다: 인덱싱과 슬라이싱이 가능하다
t1 = (1,2,"a", "b")
t2 = (3,4)
print(t1 + t2) #새로운 튜플이 만들어지고 출력된거임. 
print(t1)
print(t2)
```

```python
#튜플은 순서 바뀌는게 편한거임
a = 5
b = 3
print(a,b)
a, b = b, a
print(a,b) 
```
