```python
# f - string
# 정렬 : 왼쪽 정렬, 오른쪽 정렬, 가운데 정렬
h  = "hello"
print(h)
print(f"{h}")
print(f"{h:10}")  #10자리 생성 + 왼쪽 정렬
print(f"{h:>10}") #10자리 생성 + 오른쪽 정렬
# ->  오른쪽 정렬한다는 뜻임 
print(f"{h:^10}")
```

```python
# f-string 에서 {} 표현식이라고 한다. 중괄호 안에 식(코드) 가 들어갈 수 있다
# 조건식 부여: if 문
h = "hello"
b = "bye"
print(f"{h if len(h) > len(b) else b}")
#중괄호 {} 안에 포맷 스트링이 들어갈 수 있음
print(f"{h if len(h) < len(b) else b}")
```
