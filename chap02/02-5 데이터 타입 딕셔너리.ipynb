```python
# 딕셔너리 = 데이터 타입 -> 저장소 
# 6가지 데이터 타입 모두 저장소임
# 저장소를 변수라고 함
# 변수를 파이썬에서는 속성/변수라고 함]
# 인덱스가 있다는것은 인덱싱이나 슬라이싱을 할 수 있음

# 숫자형 : 정수, 실수 저장소
# 문자열: 종류 -> 문자열 저장소, 형태 -> 인덱스 사용, 불변 즉 값을 바꿀 수 없다
# 리스트: 종류 -> 아무거나 다 저장, 형태 -> 인덱스 사용, 가변
# 튜플: 종류 -> 아무거나 다 저장, 형태 -> 인덱스 사용, 불변 (튜플은 method자체가 존재할 수 없음 - 불변이기때문에) 불변 즉 값을 바꿀 수 없다
# 딕셔너리 : 종류 -> value는 아무거나 다 저장, 형태 -> 인덱스 없다. (인덱스가 없다는것은 인덱싱이나 슬라이싱을 사용할 수 없음)
# 딕셔너리는 value를 저장하려고 만든거임. 튜플도 마찬가지.  다만 value를 key 값으로 접근하겠다는거임
# 딕셔너리의 key는 문자열로 가는거임  value는 6가지 데이터 타입으로 모두 다 저장가능
# 딕셔너리는 key를 사용해서 데이터에 접근하게 할수있음. 불변이 아니라 가변이다
# 딕셔너리를 사용하는 이유는 리스트를 사용해서 접근하면 데이터가 더 많아지면 뭐가 뭔지 모르게 됨
# key 값은 불변이다. 
# set : 종류 -> 아무거나 다 저장, 형태 -> 인덱스랑 key 둘다 없다. 아무거나 다 저장 가능, 중복 허용 안됨
```

```python
# 딕셔너리 생성 방법: {key: value} or {key1: value1, key2: value2, ....}
# 국어점수

#리스트 처리방식
names = ["홍길동", "고길동", "마이꼴"]
kors = [80, 90, 60]
print(kors[2])

#딕셔너리 처리방식  { "key": value, "key2": value2, "key3": value3 }
# key랑 value를 한쌍으로 다룬다
#딕셔너리: 코딩이 편해짐 - key를 [key] 넣으면 value를 바로 뱉어냄
kor_dict = {
    "홍길동" : 80,
    "고길동" : 90,
    "마이꼴" : 60
}

#마이꼴이라는 key 값으로 접근하게 
print(kor_dict["고길동"])
```

```python
#딕셔너리 쌍 (key: value) 추가하기
a = {1: "a"}
print(a[1])

#key: value 쌍을 추가 2: "b"   추가하는방법
a[2]= "b"

print(a)
print(a[2])

#key따로 value 따로 분리되어있음
#추가:  name:"pey"
a["name"] = "pey"
print(a)

#추가한 "name": "pey"   접근하는방법
print(a["name"])

#추가 리스트
a[3] = [1,2,3]
print(a["name"])
```

```python
#딕셔너리 value 삭제
#object를 지울 수 있는 함수 del 을 사용해서 value를 지우기 

del a[1]
print(a)
```

```python
#딕셔너리에서 key를 사용해서 value값 얻기
# pey 점수 10, Julliet 점수 99 저장하는 딕셔너리 생성
# 딕셔너리는 {}

score = {
    "Pey" : 10,
    "Julliet" : 99
}

print(score["Julliet"])
```

```python
#딕셔너리 만들때 주의할 사항
# key 값은 중복되면 안된다

scores = [80, 90, 60]
scores_dict = {
    0:80,
    1:90,
    2:60
}
print(scores[0], scores_dict[0])
```

#### 함수와 메소드의 차이점 요약
|구분|함수(function)|메소드(method)
|----|----|----|
|소속|특정 객체에 종속되지않고 독립적으로 존재|클래스 내부에 정의되어 오브젝트에 종속
|호출방법|함수이름() 형태로 직접 호출|객체이름, 메소드이름() 형태로 객체를 통해 호출

```python
#딕셔너리 메소드 dictionary method
# method | key 만 추출하는거
# 1. keys()  <- 이 method는 key:value에서 key 부분을 반환함

#1명의 연락처 정보가 들어가있는 딕셔너리 정보
a = {
    'name': 'pey',
    'phone': '01012341234',
    'birthday': '1118'
}

#key만 뺴내고 싶을때
print(a.keys())

keys_a = a.keys()
print(keys_a, type(keys_a))
```

```python
# values 추출 method: values() => 값들을 dict_values 오브젝트에 담아서 반환
# print(a)
# b = list(a.values())
# print(b[1])

list(a.values())[0]
```

```python
# key 왜 value 한쌍으로 추출 method : items()
# (,) 튜플로 한쌍 묶여서 반환
print(a.items())
```

```python
# 모두 지우기 : clear()
# clear() 하면 {} dictionary object 자체는 살아있고 key랑 value만 삭제된거임 
print(a)
```

```python
#key로 value 얻기
a = {'name': 'pey', 'phone': '01012341234', 'birth': '1118'}
print(a)
# 'pey' 추출   name을 key로 하는 value
print(a['name']) #대괄호 연산자로 추출하는 방식
print(a.get('name')) #get이라는 method로 추출하는 방식

# 대괄호 연산자랑 get method의 차이점은?

# 대괄호 연산자는 key error가 뜸. 
# error가 나면 밑에가 아예 실행이 안됨
# 그대신 method를 사용하면 그냥 none 처리가 되서 최소한 error가 발생하지 않는다
# print(a['names'])
print(a.get('names'))

#get method 사용하는 게 낫다 

if a.get('names'):
    print('이름이 있습니다')
else:
    print('이름이 없습니다')
```

```python
# 해당 key 가 해당 dictionary안에 있는지 확인: in
print(a)
print('name' in a)
#해당 key가 있으면 true, 없으면 false 를 반환한다

if 'names' in a :
    print('names라는 key 있습니다')
else:
    print('names key 없습니다')
```
