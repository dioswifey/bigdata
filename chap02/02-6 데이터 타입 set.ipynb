```python
# set 이라는 함수를 사용, 여러개 저장
# set 은 중복된걸 제거한다
# 인덱싱이 안된다  -> 인덱스가 없다
# set은 index가 없다 -> 인덱싱이 없다
# set은 6가지의 데이터 타입 중 하나이고
# 모든 데이터 타입은 데이터 저장소의 역할을 한다


s1 = set([3,1,1,4,2,3,1])  #set 오브젝트가 생성, 중복 제거하고 생성
print(s1)
print(list(s1)[1])
```

```python
# 중복된것을 빼는데 사용하는 set
s2 = set("Hello")
print(s2)
```

```python
# 집합(set): 중복 제거, 합집합, 교집합, 차집합
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])
#교집합

#s1과 s2가 각각 고객일때 둘다 좋아하는 상품일 때
# s1 & s2
# print(s1.intersection(s2)) #method
# #합집합 결과  s1 | s2
# print(s1.union(s2))
#차집합 결과 s1  - s2
#차집함은 unique 한걸 보려고할떄   in this case, s1
s1.difference(s2)
```

```python
#set method
#add(추가), update(수정), remove(제거)

#add: 한개 추가
#update : 여러개 추가

s1 = set([1,2,3])
s1.update([4,5,6])
```
